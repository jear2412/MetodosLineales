```
#------ kNN datasets


import numpy as np
import scipy
import scipy.stats
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
import pandas as pd

sns.set_theme(style="darkgrid") 
```

```
#10 blue means
# mu_k ~ N_2( (1,0)', I   )

#blue means

bmu=scipy.stats.multivariate_normal.rvs(mean=np.array([1,0])  , cov=np.identity(2), size=10, random_state=None)

#10 orange means
# mu_k ~ N_2( (0,1)', I   )

omu=scipy.stats.multivariate_normal.rvs(mean=np.array([0,1])  , cov=np.identity(2), size=10, random_state=None)
```

```
#Medias
plt.title('Medias')
plt.plot( bmu[:,0], bmu[:,1], '*')
plt.plot( omu[:,0], omu[:,1], '*')
plt.xlabel(r'$x_1$')
plt.ylabel(r'$x_2$')
```

```
#---------- Train
#---blue observations

n=100


bobs=np.zeros((n, 2))
for i in range(n):
  k= np.random.randint(10, size=1)
  bobs[i]= scipy.stats.multivariate_normal.rvs(mean= bmu[k[0]]  , cov=np.identity(2)/5, size=1, random_state=None)

#---orange observations

n=100

oobs=np.zeros((n, 2))
for i in range(n):
  k= np.random.randint(10, size=1)
  oobs[i]= scipy.stats.multivariate_normal.rvs(mean= omu[k[0]]  , cov=np.identity(2)/5, size=1, random_state=None)

```

```
plt.title('Datos')
plt.plot( bobs[:,0], bobs[:,1], '.')
plt.plot( oobs[:,0], oobs[:,1], '.')
plt.xlabel(r'$x_1$')
plt.ylabel(r'$x_2$')
```

```
x1=np.append( bobs[:,0], oobs[:,0]    )
x2=np.append( bobs[:,1], oobs[:,1]    )
y=np.append( np.repeat(0,n), np.repeat(1,n))
```

```
train= { 'x1': x1, 'x2':x2,  'y':y }
train=pd.DataFrame(train)

train.to_csv('train.csv') 
```

```
#---------- Test
#---blue observations

n=10000


testbobs=np.zeros((n, 2))
for i in range(n):
  k= np.random.randint(10, size=1)
  testbobs[i]= scipy.stats.multivariate_normal.rvs(mean= bmu[k[0]]  , cov=np.identity(2)/5, size=1, random_state=None)

#---orange observations


testoobs=np.zeros((n, 2))
for i in range(n):
  k= np.random.randint(10, size=1)
  testoobs[i]= scipy.stats.multivariate_normal.rvs(mean= omu[k[0]]  , cov=np.identity(2)/5, size=1, random_state=None)
```

```
plt.title('Datos')
plt.plot( testbobs[:,0], testbobs[:,1], '.')
plt.plot( testoobs[:,0], testoobs[:,1], '.')
plt.xlabel(r'$x_1$')
plt.ylabel(r'$x_2$')
```

```
x1=np.append( testbobs[:,0], testoobs[:,0]    )
x2=np.append( testbobs[:,1], testoobs[:,1]    )
y=np.append( np.repeat(0,n), np.repeat(1,n))
```
