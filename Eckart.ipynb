```
#Eckart Young aplicacion

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import cv2
```

```
img = cv2.imread('gabe.jpg', 0)

# obtain svd
U, S, V = np.linalg.svd(img)

```

```
print(S)
```

```
print(img)
```

```
# inspect shapes of the matrices
print(U.shape, S.shape, V.shape)
```

```
np.sum(S[0:100])/np.sum(S)
```

```
# plot images with different number of components
comps = [10, 15, 20, 30, 40, 100]

plt.figure(figsize = (16, 8))
for i in range(6):
  low_rank = U[:, :comps[i]] @ np.diag(S[:comps[i]]) @ V[:comps[i], :]
  if(i  == 0):
     plt.subplot(2, 3, i+1), plt.imshow(low_rank, cmap = 'gray'), plt.axis('off'), plt.title("Imagen original con componentes =" + str(comps[i]))
  else:
     plt.subplot(2, 3, i+1), plt.imshow(low_rank, cmap = 'gray'), plt.axis('off'), plt.title("componentes =" + str(comps[i]))
```
